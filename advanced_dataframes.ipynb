```python
mpg = data('mpg')
```

```python
mpg.head()
```

```python
mpg.info()

# 234 rows, 11 columns
```

```python
mpg = mpg.rename(columns= {'cty': 'city', 'hwy': 'highway'})
mpg.head()
```

```python
mpg.describe()
```

```python
len(mpg.manufacturer.value_counts())

# 15 different manufacturers
```

```python
mpg.model.value_counts().size

# 38 different models
```

```python
mpg['mileage_difference'] = mpg.highway - mpg.city
mpg
```

```python
mpg['average_mileage'] = (mpg.city + mpg.highway)/2
mpg
```

```python
mpg['is_automatic'] = np.where(mpg.trans.str.startswith('auto'), 'automatic', 'manual')
mpg
```

```python
# Using the mpg dataset, find out which which manufacturer has the best miles per gallon on average.
best_mpg_by_brand = mpg.groupby('manufacturer').average_mileage.max()
type(best_mpg_by_brand)
best_mpg_by_brand[best_mpg_by_brand == best_mpg_by_brand.max()]

# Does Volkswagen havethe best average mpg? No, Volkswagen makes the car with the highest mpg.
```

```python
avg_mpg_by_mfr = mpg.groupby('manufacturer').average_mileage.mean()
type(avg_mpg_by_mfr)
best_mpg_by_brand = avg_mpg_by_mfr[avg_mpg_by_mfr == avg_mpg_by_mfr.max()]
best_mpg_by_brand

# Honda has the best avg mpg.
```

```python
# Do automatic or manual cars have better mpg?

best_mpg_by_transmission = mpg.groupby('is_automatic').average_mileage.mean()
type(best_mpg_by_transmission)
best_mpg_by_transmission[best_mpg_by_transmission == best_mpg_by_transmission.max()]

#Manual transmissions have better mpg than automatics.
```

```python
url = get_db_url('chipotle')
chipotle_table = pd.read_sql('SELECT * FROM orders', url)
```

```python
pd.read_sql('SHOW TABLES', url)
```

```python
chipotle_table.info()
```

```python
chipotle_table.head()
```

```python
chipotle_table['item_price_flt'] = chipotle_table['item_price'].str.lstrip('$').astype(float)
chipotle_table.head()
```

```python
# What is total price for each order?
chipotle_table['order_subtotal'] = chipotle_table.quantity * chipotle_table.item_price_flt
chipotle_table.head()

# See next cell.
```

```python
order_total = chipotle_table.groupby('order_id').order_subtotal.sum()
type(order_total)
chipotle_table['order_total'] = order_total
chipotle_table.head()

# order_total series lists total price for each order
```

```python
# What are the most popular 3 items?

chipotle_table['item_name'].value_counts().head(3)
```

```python
# Which item has produced the most revenue?

total_revenue = chipotle_table.groupby('item_name').order_total.sum()
total_revenue[total_revenue == total_revenue.max()]

# The chicken burrito has produced the most revenue
```

```python
# Write the code necessary to create a cross tabulation of the number of titles by department.

dept_emp_titles = pd.read_sql('''
    SELECT dept_emp.dept_no, titles.title, departments.dept_name 
    FROM titles 
    JOIN dept_emp ON titles.emp_no = dept_emp.emp_no 
    JOIN departments ON departments.dept_no = dept_emp.dept_no
    ''', url)
dept_emp_titles.head()
```

```python
pd.crosstab(dept_emp_titles.title, dept_emp_titles.dept_name, margins= True)

# I might be proud of myself if this is right...
```
