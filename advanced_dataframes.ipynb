```python
users = pd.DataFrame({
    'id': [1, 2, 3, 4, 5, 6],
    'name': ['bob', 'joe', 'sally', 'adam', 'jane', 'mike'],
    'role_id': [1, 2, 3, 3, np.nan, np.nan]
})
users
```

```python
roles = pd.DataFrame({
    'id': [1, 2, 3, 4],
    'name': ['admin', 'author', 'reviewer', 'commenter']
})
roles
```

```python
pd.merge(users, roles, on='id', how='right')

# Users Jane and Mike, with null values for them in the role_id and name columns, are lost in a right join.
```

```python
pd.merge(users, roles, on= 'id', how= 'outer')

# All users are included.
```

```python
users_fkd = users.drop(columns= 'id')
users_fkd
```

```python
roles_fkd =users.drop(columns= 'id')
roles_fkd
```

```python
pd.merge(users_fkd, roles_fkd, how= 'inner')

# Identical dataframe produced
```

```python
mpg = data('mpg')
mpg.head()
```

```python
mpg.info()

# 234 rows, 11 columns
```

```python
mpg = mpg.rename(columns= {'cty': 'city', 'hwy': 'highway'})
mpg.head()
```

```python
mpg['mileage_difference'] = mpg.highway - mpg.city
mpg
```

```python
mpg['average_mileage'] = (mpg.city + mpg.highway)/2
mpg
```

```python
mpg['is_automatic'] = np.where(mpg.trans.str.startswith('auto'), 'automatic', 'manual')
mpg
```

```python
# Using the mpg dataset, find out which which manufacturer has the best miles per gallon on average.
best_mpg_by_brand = mpg.groupby('manufacturer').average_mileage.max()
type(best_mpg_by_brand)
best_mpg_by_brand[best_mpg_by_brand == best_mpg_by_brand.max()]

# Does Volkswagen havethe best average mpg? No, Volkswagen makes the car with the highest mpg.
```

```python
avg_mpg_by_mfr = mpg.groupby('manufacturer').average_mileage.mean()
type(avg_mpg_by_mfr)
best_mpg_by_brand = avg_mpg_by_mfr[avg_mpg_by_mfr == avg_mpg_by_mfr.max()]

# Honda has the best avg mpg.
```

```python
# Do automatic or manual cars have better mpg?

best_mpg_by_transmission = mpg.groupby('is_automatic').average_mileage.mean()
type(best_mpg_by_transmission)
best_mpg_by_transmission[best_mpg_by_transmission == best_mpg_by_transmission.max()]

#Manual transmissions have better mpg than automatics.
```
