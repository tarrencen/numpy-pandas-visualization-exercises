1. Copy the code from the lesson to create a dataframe full of student grades.

    a. Create a column named passing_english that indicates whether each student has a passing grade in english.

    b. Sort the english grades by the passing_english column. How are duplicates handled?

    c. Sort the english grades first by passing_english and then by student name. All the students that are failing english should be first, and within the students that are failing english they should be ordered alphabetically. The same should be true for the students passing english. (Hint: you can pass a list to the .sort_values method)

    d. Sort the english grades first by passing_english, and then by the actual english grade, similar to how we did in the last step.
    
    e. Calculate each students overall grade and add it as a column on the dataframe. The overall grade is the average of the math, english, and reading grades.

```python
import pandas as pd
from pydataset import data
import numpy as np

np.random.seed(724)

students = ['Sally', 'Jane', 'Suzie', 'Billy', 'Ada', 'John', 'Thomas',
            'Marie', 'Albert', 'Richard', 'Isaac', 'Alan']

# randomly generate scores for each student for each subject
# note that all the values need to have the same length here
math_grades = np.random.randint(low=60, high=100, size=len(students))
english_grades = np.random.randint(low=60, high=100, size=len(students))
reading_grades = np.random.randint(low=60, high=100, size=len(students))

df = pd.DataFrame({'name': students,
                   'math': math_grades,
                   'english': english_grades,
                   'reading': reading_grades})

type(df)
```

```python
print(df)
```

```python
df.info()
```

```python
df.describe()
```

```python
# Exercise 1a - Create a column named passing_english that indicates whether each student has a passing grade in English.

df['passing_english'] = df.english >= 70
df
```

```python
# Exercise 1b - Sort the English grades by the passing_english column. How are duplicates handled?

df[df.english > 0].sort_values(by='passing_english')

# What duplicates?
```

```python
# Exercise 1c - Sort the english grades first by passing_english and then by student name. All the students that are failing english should be first, and within the students
# that are failing english they should be ordered alphabetically. The same should be true for the students passing english. (Hint: you can pass a list to the .sort_values
# method)

df.sort_values(by=['passing_english', 'name'], ascending=[True, True])
```

```python
# Exercise 1d - Sort the english grades first by passing_english, and then by the actual english grade, similar to how we did in the last step.

df.sort_values(by=['passing_english', 'english'], ascending=[True, False])
```

```python
# Exercise 1e - Calculate each students overall grade and add it as a column on the dataframe. The overall grade is the average of the math, english, and reading grades.

df['gpa'] = round((df.math + df.english + df.reading)/3, 2)
df

# Was that the dumb way? Is there a more programmatic method?
```

```python
# Exercises 2d, 2e -

mpg = mpg.rename(columns={'cty': 'city', 'hwy': 'highway'})
```

```python
# Exercise 2f - Do any cars have better city mileage than highway mileage?

mpg[mpg.city > mpg.highway]

# No, there are no cars with better city mileage than highway mileage.
```

```python
# Exercises 2g - Create a column named mileage_difference this column should contain the difference between highway and city mileage for each car.

mpg['mileage_difference'] = mpg.highway - mpg.city
mpg.drop(columns= {'average_mileage'})


```

```python
# Exercise 2h - Which car (or cars) has the highest mileage difference?

mpg[mpg.mileage_difference == mpg.mileage_difference.max()]


# Honda Civic and Volkswagen New Beetle both have a mileage difference of 12 mpg.
```

```python
mpg = mpg.drop(columns= 'mileage_diff')
```

```python
# Exercise 2i part 1- Which compact class car has the lowest highway mileage? 
compact_cars = mpg[mpg['class'] == 'compact']
compact_cars[compact_cars.hwy == compact_cars.hwy.min()]
```

```python
# Exercise 2i part 2 - The best?

compact_cars[compact_cars.hwy == compact_cars.hwy.max()]
```

```python
# Exercise 2j - Create a column named average_mileage that is the mean of the city and highway mileage.
mpg['average_mileage'] = (mpg.city + mpg.highway)/2
mpg.head()
```

```python
# Exercise 2k part 1 - Which Dodge car has the best average mileage?

dodge_cars = mpg.manufacturer == 'dodge'
mpg[dodge_cars].sort_values(by= 'average_mileage', ascending= False).head(1)
```

```python
# Exercise 2k part 2 - The worst?

mpg[dodge_cars].sort_values(by= 'average_mileage').head(1)
```

3. Load the Mammals dataset. Read the documentation for it, and use the data to answer these questions:

    a. How many rows and columns are there?

    b. What are the data types?

    c. Summarize the dataframe with .info and .describe

    d. What is the the weight of the fastest animal?

    e. What is the overal percentage of specials?
    
    f. How many animals are hoppers that are above the median speed? What percentage is this?

```python
# Exercises 3a, 3b, 3c - How many rows and columns are there?

mammals = data('Mammals')

mammals.shape
#107 rows, 4 columns

mammals.info()
# Floats and booleans are the data types

mammals.describe()
# 2 quantitative columns, weight and speed

```

```python
# Exercise 3d - What is the weight of the fastest animal?

mammals.sort_values(by= ['speed', 'weight'], ascending= [False, True]).head(1)

# The weight of the fastest animal is 55.0 (kgs?)
```

```python
# Exercise 3e - What is the overall percentage of specials?
round((len(mammals[mammals.specials == True])/len(mammals)) * 100, 2)

# 9.35% 
```

```python
# Exercise 3f part 1 - How many animals are hoppers that are above the median speed? 
hoppers = mammals[mammals.hoppers == True]
mammals.speed.median()
fast_hoppers = hoppers[hoppers > mammals.speed.median()]
len(fast_hoppers)
```

```python
# Exercise 3f part 2 - What percentage is this?

round(len(fast_hoppers)/len(mammals) * 100, 2)

# 10.28%
```
