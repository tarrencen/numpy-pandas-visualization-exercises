#Use matplotlib to plot the following equation:

#y = x^2 - x + 2

```python
import matplotlib.pyplot as plt
import math
import numpy as np

%matplotlib inline
x = range(-50,50)
y = [((n ** 2) - n + 2) for n in x]

plt.plot(x, y)
plt.text(3.5,-10, 'origin', fontsize = 10, color = '#333333')
```

# Create and label 4 separate charts for the following equations (choose a range for x that makes sense):

#y = x ** .05
#y = x ** 3
#y = 2 ** x
#y = 1/(x + 1)

```python
x = range(0,64)

y = [n ** 0.5 for n in x]

plt.plot(x, y)

plt.text(0.1,0, 'origin', fontsize = 10, color = '#444444')
```

```python
x = range(-8,8)
y = [n ** 3 for n in x]

plt.plot(x, y)
plt.text(0.1,-15, 'origin', fontsize = 10, color = 'r')
plt.title('$y = x^3$', fontsize = 20)
plt.xlabel('$x$', fontsize = 20)
plt.ylabel('$x^3$', fontsize = 20)
```

```python
x = np.linspace(0,10)
y = [2 ** n for n in x]


plt.plot(x, y)
plt.text(0.25,0, 'origin', fontsize = 10, color = 'r')
plt.title('$y = 2^x$', fontsize = 20)
plt.xlabel('$x$', fontsize = 20)
plt.ylabel('$2^x$', fontsize = 20)


```

```python
x = range(1,10)
y = [(1/(n + 1)) for n in x]

plt.plot(x,y)

plt.text(0,0, 'origin', fontsize = 10, color = 'purple')
plt.title('$y = 1/(x + 1)$', fontsize = 20)
plt.xlabel('$x$', fontsize = 20)
plt.ylabel('$1/(x + 1)$', fontsize = 20)
```
